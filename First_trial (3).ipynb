###Handle outliers

```python
# Filter the dataframe for Churn == 'Yes'
churn_yes = df[df['Churn'] == 'Yes']

# Calculate the IQR
Q1 = churn_yes['tenure'].quantile(0.25)
Q3 = churn_yes['tenure'].quantile(0.75)
IQR = Q3 - Q1

# Define margins for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Find out the number of outliers
outliers_low = churn_yes[churn_yes['tenure'] < lower_bound]
outliers_high = churn_yes[churn_yes['tenure'] > upper_bound]

# count the number of outliers
num_outliers = len(outliers_low) + len(outliers_high)
print('Number of outliers:', num_outliers)
```

###Encoding categorical variable

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
columns_to_encode = ['MultipleLines', 'InternetService', 'OnlineSecurity', 'OnlineBackup',
                     'DeviceProtection', 'TechSupport', 'Contract', 'StreamingTV',
                     'StreamingMovies', 'PaymentMethod']

# Create the OneHotEncoder object
encoder = OneHotEncoder(sparse=False, handle_unknown='ignore')  # handle_unknown='ignore' to deal with unknown categories in the test set

# Fit the encoder on the training data
encoder.fit(X_train[columns_to_encode])

# Transform both training and test data
X_train_encoded = encoder.transform(X_train[columns_to_encode])
X_test_encoded = encoder.transform(X_test[columns_to_encode])

# Create DataFrames from the encoded arrays
X_train_encoded_df = pd.DataFrame(X_train_encoded, columns=encoder.get_feature_names_out(columns_to_encode), index=X_train.index)
X_test_encoded_df = pd.DataFrame(X_test_encoded, columns=encoder.get_feature_names_out(columns_to_encode), index=X_test.index)

# Drop the original columns and concatenate the new one-hot encoded columns for both training and test sets
X_train = X_train.drop(columns_to_encode, axis=1).reset_index(drop=True)
X_test = X_test.drop(columns_to_encode, axis=1).reset_index(drop=True)

X_train = pd.concat([X_train, X_train_encoded_df], axis=1)
X_test = pd.concat([X_test, X_test_encoded_df], axis=1)
```

```python
for column in X_train.columns:
    if X_train[column].dtype == 'float64':
        X_train[column] = X_train[column].astype(int)
```

```python
for column in X_test.columns:
    if X_test[column].dtype == 'float64':
        X_test[column] = X_test[column].astype(int)
```

#UNIVARIATE ANALYSIS

```python
import pandas as pd
from tqdm import tqdm

def univariate_analysis_categorical(df, column):
    # Calculate frequency counts and percentages
    frequency = df[column].value_counts()
    percentage = df[column].value_counts(normalize=True) * 100
    return frequency, percentage

def univariate_analysis_numerical(df, column):
    # Calculate summary statistics
    summary = df[column].describe()
    return summary

categorical_columns = ['gender', 'SeniorCitizen', 'Partner', 'Dependents', 'PhoneService', 'MultipleLines',
                       'InternetService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport',
                       'StreamingTV', 'StreamingMovies']

# Analyze categorical variables
for column in tqdm(categorical_columns, desc='Analyzing Categorical Variables'):
    freq, perc = univariate_analysis_categorical(df, column)
    print(f'\n{column} - Frequency:\n{freq}')
    print(f'{column} - Percentage:\n{perc}')

# Analyze numerical variable
print('\nNumerical Variable - Tenure Summary Statistics:')
print(univariate_analysis_numerical(df, 'tenure'))
```
